```python
import math
print("10 Tabanında Log,Mutlak Değer ve Karekök Bulma Sistemine Hoşgeldiniz.")
x = float(input("Bir Sayı Giriniz"))
dosya = open("sonuçlar.txt", mode="a")
a = math.log(x, 10)
b = math.sqrt(x)
c = math.fabs(x)
dosya.write(f"Sayınızın 10 Tabanında Logaritması: {a}\n")
dosya.write(f"Sayınızın Karekökü: {b}\n")
dosya.write(f"Sayınızın Mutlak Değeri: {c}\n")
dosya.close()
dosya = open("sonuçlar.txt", mode="r")
print(dosya.readline())
print(dosya.readline())
print(dosya.readline())
dosya.close()
```

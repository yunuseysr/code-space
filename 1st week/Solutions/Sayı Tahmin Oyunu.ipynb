```python
import random

x = random.randint(1, 101)
while True:
    veri = int(input("0-100 Arasında Sayı Tuttum.Sence Bu Sayı Nedir?"))
    if x == veri:
        print("Doğru Tahmin.")
        break
    elif x < veri:
        print("Daha Ufak Bir Sayı Tahmin Ediniz")
        continue
    elif veri < x:
        print("Daha Büyük Bir Sayı Tahmin Ediniz.")
        continue
```

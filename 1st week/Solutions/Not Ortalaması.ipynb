```python
not1 = input('İlk Sınav Notunuzu Giriniz:')
not2 = input('İkinci Sınav Notunuzu Giriniz:')
ilknot = int(not1)
ikincinot = int(not2)
ortalama = (ilknot + ikincinot)/2
if ortalama < 50 :
    print('Kaldınız.')
elif 50 <= ortalama <= 60:
    print('Geçtin.')
elif 60 < ortalama <= 80:
    print('İyi bir ortalama ile geçtin.')
elif 80 < ortalama:
    print('Çok iyi bir ortalama ile geçtin.')
```
